```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
for N in tqdm([1,10,100, 1000, 100000000]):
    

    x = np.random.randint(1, 7, size=N)
    y = np.random.randint(1, 7, size=N)

    sum = x+y

    plt.figure(figsize=(6, 4))  # create a new figure
    plt.hist(sum, bins=50, density=False, label='original data', color = 'b', edgecolor='black')
    plt.show
```
