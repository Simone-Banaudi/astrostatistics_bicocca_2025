## Poisson distribution

```python
k = np.array([0,1,2,3,4]) # k of the poisson distribution

for mu in [0.1,0.5,1,2]:
    p = scipy.stats.poisson(mu).pmf(k) # pmf(k, mu, loc=0) Probability Mass Function, here mu is frozen in poisson(mu).
    plt.plot(k, p,label=fr'$\mu='+str(mu)+"$")
plt.legend()

plt.xlabel("k")  # X-axis label
plt.ylabel("Probability Mass Function (PMF)")  # Y-axis label
plt.title(r"Poisson Distribution for Different $\mu$ Values")  # Title
```

## Boolean mask (an example with rejection sampling)

```python
f = lambda x: x+np.sin(x)  # pause and look at how great lambda functions are! 

N=1000

ymax = 10
x = np.random.uniform(0,10,N)
y = np.random.uniform(0,ymax,N)

goodpoints = x[y<f(x)] # BOOLEAN MASK: each x is associated with an y

plt.hist(goodpoints,bins=50,density=True,histtype='step',color='green');
```
